# Streaming and Batch

### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling `stream()` returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
# model gives output when full text gererated.
model.invoke("Write a 500 words paragraph on AI")
```

```python
# Now Applying streaming --> Streaming means the model shows the response little by little as it generates it instead of all at once.
model.stream("Write a 500 words paragraph on AI")

# --> tihs line of code creates a 'generator object'  ---> iterator
```

```python
type(model.stream("Write a 500 words paragraph on AI"))
```

```python
for chunk in model.stream("Write a 500 words paragraph on AI"):
    print(chunk.text,end='',flush=True)
```

### Batch
Batching a collection of independent requests to a model can significantly improve performance and reduce costs,as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "Why do parrots have colorful featurs?",
        "How do airplanes fly?",
        "What is Quantum computing?"
    ]
)

for response in responses:
    print(response)
    print("="*50)
```

```python
# Limit to 5 parallel calls

# Example: If 6 questions are given and max_concurrency = 5

# First 5 run simultaneously

# 6th runs after one finishes.

# Note : max_concurrency = parallel requests, not total requests.

responses = model.batch(
    [
        "Why do parrots have colorful featurs?",
        "How do airplanes fly?",
        "What is Quantum computing?",
        "What is AI?",
        "What is Machine Learning?",
        "What is Deep Learning?"
    ],
    config={
        "max_concurrency": 5
    }
)

for response in responses:
    print(response)
    print("="*50)
```
