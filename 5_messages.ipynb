# Messages

Messages are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:
   - **Role** - Identifies the message type (e.g. system, user)
   - **Content** - Represents the actual content of the message (like text, images, audio, documents, etc.)
   - **Metadata** - Optional fields such as response information, message IDs, and token usage

LangChain provides a standard message type that works across all model providers, ensuring consistent behavior regardless of the model being called.

# Message Prompts
- Alternatively, you can pass in a list of messages to the model by providing a list of message objects.

### Message types

- *System message* - Tells the model how to behave and provide context for interactions
- *Human message* - Represents user input and interactions with the model
- *AI message* - Responses generated by the model, including text content, tool calls, and metadata
- *Tool message* - Represents the outputs of tool calls

---
`1. System Message :`
- A SystemMessage represent an initial set of instructions that primes the model’s behavior. You can use a system message to set the tone, define the model’s role, and establish guidelines for responses.

`2. Human Message :`
- A HumanMessage represents user input and interactions. They can contain text, images, audio, files, and any other amount of multimodal content.

`3. AI Message :`
- An AIMessage represents the output of a model invocation. They can include multimodal data, tool calls, and provider-specific metadata that you can later access.

`4. Tool Message :`
- For models that support tool calling, AI messages can contain tool calls. Tool messages are used to pass the results of a single tool execution back to the mode 

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage   

messages = [
    SystemMessage("You are a doctor"),
    HumanMessage("Tell me the symptoms of Diabetic Retinopathy")
]

response = model.invoke(messages)
print(response.content)
```

### Message Metadata in LangChain

**Message metadata** is **extra information attached to a message** that is not part of the main text (`content`) but helps track how the message was generated.

`# Purpose`
- Stores **technical details about the message**
- Helps with **tool calling, debugging, tracing, and logging**
- Provides **information about the model's behavior**

`# Common Metadata Fields`

Some common metadata fields include:

- **tool_calls** → Tools that the model decided to call  
- **id** → Unique identifier for the message  
- **model_name** → Which model generated the response  
- **token_usage** → Number of tokens used in the request/response  
- **finish_reason** → Why the model stopped generating  

`# Example AIMessage`

AIMessage(
content="It's sunny in New York",
metadata={
"model_name": "llama-3.1-8b-instant",
"token_usage": {
"prompt_tokens": 20,
"completion_tokens": 10
}
}
)


`# Structure`

Message =  
- **content** → Actual response text  
- **metadata** → Extra technical information

`# Key Idea`
- Metadata helps developers **understand how the LLM generated the response**, track **tool usage**, and monitor **model performance**.

```python
## Message Metadata
human_msg = HumanMessage(
    content="Hello!",
    name="alice",  # Optional: identify different users
    id="msg_123",  # Optional: unique identifier for tracing
)
```

```python
## AIMessage Tool Call Example (LangChain)
from langchain.messages import AIMessage,ToolMessage

# After a model makes a tool call 
# Here , we demonstrate manually creating the message for brevity)
ai_message = AIMessage(
    content = [],
    tool_calls = [{
        "name":"get_weather",
        "args":{"location":"Pune"},
        "id":"call_123"
    }]
)
```

### AIMessage Tool Call Example (LangChain)

- This snippet shows how to **manually create an AIMessage that contains a tool call**.
- from langchain.messages import AIMessage, ToolMessage
- After the model decides a tool is required, it generates an **AIMessage with tool call information**.

ai_message = AIMessage(
    content = [],
    tool_calls = [{
        "name": "get_weather",
        "args": {"location": "Pune"},
        "id": "call_123"
    }]
)

`# Explanation :`

- **AIMessage** → Represents the response generated by the LLM.
- **content = []** → No direct text response because the model wants to call a tool.
- **tool_calls** → Contains the details of the tool the model wants to use.

`# Tool Call Fields : `

- **name** → The tool to execute (`get_weather`)
- **args** → Arguments passed to the tool (`location = Pune`)
- **id** → Unique identifier for this tool call (`call_123`)

`# Key Idea :`

- Instead of answering directly, the **LLM asks the system to execute a tool**.  
- The tool result will later be returned using a **ToolMessage**, and then the model generates the final response.

```python
# Execute tool and create result message
weather_result = "Sunny ,72°F"
tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "call_123"  # Must match the call_ID
)
```

###  ToolMessage Example (LangChain)

- This snippet shows how to **create a ToolMessage after executing a tool**.
- weather_result = "Sunny, 72°F"

tool_message = ToolMessage(
    content = weather_result,
    tool_call_id = "call_123"
)

`# Explanation :`

- **weather_result**  
  → Output returned by the tool (e.g., weather API or function).

- **ToolMessage**  
  → Represents the **result produced by the tool execution**.

`# Parameters :`

- **content**  
  → The **actual result returned by the tool** (`Sunny, 72°F`).

- **tool_call_id**  
  → The **ID of the tool call generated earlier by the LLM**.  
  → It must **match the ID from AIMessage.tool_calls**.

Example earlier tool call:

{
"name": "get_weather",
"args": {"location": "Pune"},
"id": "call_123"
}

`# Matching ID is Important :`

- The model uses the **tool_call_id** to know **which tool call this result belongs to**.

`# Tool Calling Flow :`

User Question  
↓  
LLM creates tool call (AIMessage)  
↓  
Tool executes  
↓  
ToolMessage returns result  
↓  
LLM uses result to generate final answer

```python
# Continue the conversation
messages = [
    HumanMessage("What is the weathe in Pune?"),
    ai_message, # model's tool call
    tool_message # tool execution result
]

response = model.invoke(messages)  # Model processes the result
```

### Continue Conversation with Tool Results (LangChain)

messages = [
    HumanMessage("What is the weather in Pune?"),
    ai_message,      # Model's tool call
    tool_message     # Tool execution result
]

response = model.invoke(messages)

`# Explanation :`

- **HumanMessage**  
  → Represents the **user's original question**.

- **ai_message**  
  → Contains the **tool call generated by the model** (e.g., calling `get_weather`).

- **tool_message**  
  → Contains the **result returned from the tool execution**.

`# What Happens Here :`

1. The **conversation history** is stored in the `messages` list.
2. It includes:
   - User question
   - Model tool call
   - Tool execution result
3. The **model receives this full context**.
4. The model then **uses the tool result to generate the final response**.

`# Example Final Output :`
- The model may produce a response like:
- The weather in Pune is Sunny, 72°F.

`# Key Idea :`
- By sending **tool results back to the model**, the LLM can generate **accurate answers using real data instead of hallucinating**.
